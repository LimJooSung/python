#문제 2번

다형성에 대해 설명하고 다형성을 보여주는 자신만의 파이썬 코드 예제를 제시하시오. 

다형성이란?

- 상속 관계 내의 다른 클래스들의 인스턴스들이 같은 멤버 함수 호출에 대해 각각 다르게 반응하도록 하는 기능을 말한다.

- 예를들어 연산자 오버로딩도 다형성을 지원하는 중요한 기술 중의 하나라고 볼 수 있다.

다형성의 장점

- 하나의 이름 (적은 코딩)으로 다양한 객체들에게 유사한 작업을 수행시킬 수 있음 
    
- 프로그램 작성 코드 량이 줄어든다. 
    
- 코드의 가독성을 높혀준다. 
    
파이썬에서 다형성의 장점

- 형 선언이 없다는 점에서 파이썬에서는 다형성의 장점이 더욱 부각된다. 
    
- 실시간으로 객체의 형이 결정되므로 단 하나의 메소드에 의해 처리될 수 있는 객체의 수는 더욱 많아진다.
    
- 즉, 다른 언어보다 코드의 양이 더욱 줄어든다. 
    
- 파이썬에서 모든 함수는 기본적으로 다형성의 특징을 포함한다. 

```python
class Employee :
    def line(self) :
        print '아무 라인 아님'
        
class OchajangLine(Employee) :
    def line(self) :
        print '오차장 라인'
        
class ChoijeonmooLine(Employee) :
    def line(self) :
        print '최전무 라인'
        
class NoLine(Employee) :
    def line(self) :
        print '신입 or 인턴 or 사장'

class NoLine2(Employee) :
    pass

employee = []

employee.append(Employee())
employee.append(OchajangLine())
employee.append(ChoijeonmooLine())
employee.append(NoLine())
employee.append(NoLine2())

for x in range(len(employee)) :
    employee[x].line()
```

슈퍼클래스인 Employee, 서브클래스인 OchajangLine, ChoijeonmooLine, NoLine이 있다.

각 서브클래스들은 오버로딩된 line함수를 가지고 있으며 출력시 각 서브클래스에 정의된 line함수의 결과를 출력한다.

서브클래스중 내용을 가지고 있지않은 NoLine2 클래스의 경우에는 line() 출력시 슈퍼클래스의 line()함수의 결과를 출력한다.

#문제 4번

다음은 내장 자료형 list를 서브클래싱하여 만든 MySet 클래스 정의 내용이다. 다음 클래스 정의에서 __init__(), __str()__(), elimicate_duplicate()의 세 개의 메소드 코드 내용을 자신이 다른 사람에게 가르친다고 생각하며 설명해보시오.

- MySet은 집합(Set) 자료형을 정의하려는 의도하에 만들어진 클래스이다. 

```python
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + ", "
        result = result[0:len(result)-2] + "}"
        return result

    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)

    
if __name__ == "__main__":
    s = MySet([1, 2, 2, 3])
    print s
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    print t
```

현재 주어진 MySet 클래스는 서브클래스로 list를 상속받고 있다. 따라서 리스트를 따로 선언할 필요 없이 리스트 내장함수 사용이 가능하다.

전체적인 코드의 순서를 보면 __init__ -> eliminate_duplicate() -> __str__이다.

-우선 생성자 함수를 보면. self는 자바에서의 this와 같은 개념으로 인스턴스 객체 자신의 레퍼런스를 지니고 있다. 각 인스턴스들은 이 self를 이용하여

자신의 이름공간에 접근할 수 있다. l이 생성자함수로 전달되고 반복문을 돌면서 self에 각 원소를 append시킨다.

그 후 반복되는 원소를 제거하기 위하여 MySet.eliminate_duplicate(self)를 실행한다.

-다음으로 print 예약어나 str() 내장함수 호출에 대응되는 __str__함수를 보면. result = "MySet: {"으로 문자열을 생성한다.

그리고 앞에서 저장된 self를 반복문을 돌려서 문자열 result에 더해준다. 반복문이 끝나고 최종적으로 result에서 ' ,'부분을 제거하기 위하여

result = result[0:len(result)-2] + "}"처리를 해준다.(슬라이싱)

-eliminate_duplicate() 함수를 보면. 생성자에서 전달된 self를 l로 받아서 반복문을 돌면서 e가 s에 없으면 s에 e를 append시킨다.

리스트 l을 ㅣ[:]를 이용하여 빈리스트로 만들고(l이 전달되어야하므로 결과값이 l로 새로 저장되어야 함) 반복문을 돌면서 리스트 s에 저장된 값을

리스트 l에 append시킨다. 

이 과정을 거쳐서 s, t의 합집합을 출력할 수 있다.

#문제 5번

4번 문제에 정의된 MySet 클래스에 메소드를 추가하여 다음 각 요구사항 모두를 만족시키는 코딩을 제시하시오 (정답을 각 요구사항별로 입력할 필요 없이 5번 문제에 대해 1개의 MySet 클래스 정의 코드를 제시하면 된다.)

```python
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + ", "
        result = result[0:len(result)-2] + "}"
        return result

    def __or__(self, another) :    # 요구사항 1. | 연산으로 두 집합의 합집합을 반환한다.       
        tmpList = []     # 합집합을 저장할 빈 리스트
        
        for x in self :
            tmpList.append(x)
          
        for y in another :
            if y not in tmpList :
                tmpList.append(y)
                
        tmpList.sort()
                
        result = "MySet: {"
        for e in tmpList:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        
        return result
    
    def __and__(self, another) :   # 요구사항 2. & 연산으로 두 집합의 교집합을 반환한다.      
        tmpList = []       # 교집합을 저장할 빈 리스트
        
        for x in self :
            if (x in another) :
                tmpList.append(x)
         
        tmpList.sort()
        
        result = "MySet: {"
        for e in tmpList :
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        
        return result
    
    def __sub__(self, another) :   # 요구사항 3. - 연산으로 두 집합의 차집합을 반환한다.       
        tmpList = []       # 차집합을 저장할 빈 리스트
        lenS = len(self)   # self의 길이를 lenS에 저장
        
        for x in self :
            if (x not in another) :
                tmpList.append(x)
        
        tmpList.sort()
        
        result = "MySet: {"
        for e in tmpList :
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        
        return result
    
    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)

    
if __name__ == "__main__":
    s = MySet([1, 2, 2, 3])
    print s
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    print t
    
    print
    print

    s = s | t
    print " 요구사항 1 (합집합) "
    print "---------------------"
    print s
    
    print
    print
    
    s = MySet([1, 2, 3])
    t = MySet([3, 4, 5])
    s = s & t
    print " 요구사항 2 (교집합) "
    print "---------------------"
    print s
    
    print
    print
    
    s = MySet([1, 2, 3])
    t = MySet([3, 4, 5])
    s = s - t
    print " 요구사항 3 (차집합) "
    print "---------------------"
    print s
```

합집합, 교집합, 차집합을 구하는 메소드를 클래스 내부에 구현한다.

결과값을 저장할 tmpList를 생성한 후, 각각의 조건에 맞는 원소들을 tmpList에 append시킨다.

결과가 tmpList에 저장되면 tmpList.sort()를 통해 원소값을 정렬한 후에 출력되도록 한다.
